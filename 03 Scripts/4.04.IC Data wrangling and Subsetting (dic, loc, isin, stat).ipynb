### 7. Determinating the meaning behind a value of 4 in the "department_id" column within the df_prod dataframe using a data dictionary.

#### --creating the data dictionary

```python
df_dept=pd.read_csv(os.path.join(path,'02 Data','Original Data','departments.csv'), index_col=False)
```

```python
df_dept
```

#### -- Transposing the data

```python
df_dept.T
```

#### - Let's the transposed data became a new df

```python
df_dept_t=df_dept.T
```

```python
df_dept_t.head(5)
```

#### - Let's give to new df the indexes

```python
df_dept_t.reset_index().head(5)
```

#### - Copying a new header

```python
new_header=df_dept_t.iloc[0]
```

```python
new_header
```

#### - Copying without 1st row in another dataframe

```python
df_dept_t_wo_row=df_dept_t[1:]
```

```python
df_dept_t_wo_row
```

#### - Adding headers that was copyid as 'new header'

```python
df_dept_t_wo_row.columns=new_header
```

```python
df_dept_t_wo_row.head(5)
```

#### - Creating the Data dictionary 

```python
data_dict=df_dept_t_wo_row.to_dict('index')
```

```python
data_dict
```

#### - Extract department_id =4 meaning from Data dict

```python
print(data_dict.get('4'))
```
